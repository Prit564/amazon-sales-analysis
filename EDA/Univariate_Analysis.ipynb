```python
import pandas as pd
df = pd.read_csv(r"cleaned_data.csv")
df.head()
```

```python
df.describe()
```

The dataset shows that discounted prices (mean ≈ 3126) are significantly lower than actual prices (mean ≈ 5447), with an average discount of about 47.7%. Ratings are generally high (around 4.1), indicating good customer satisfaction. However, large standard deviations and high maximum values suggest the presence of strong outliers, especially in price and rating count.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['discounted_price'], bins=30)
plt.title("Distribution of Discounted Price")
plt.xlabel("Discounted Price")
plt.ylabel("Count")
```

Observation: Most products are cheap; most products are concentrated in the lower price range, indicating a right skewed distribution

```python
sns.histplot(df['rating'], bins=30)
plt.title("Distribution of rating")
plt.xlabel("Rating")
plt.ylabel("Frequency")
```

Observation: Most products rating is arount 4.2 in a range from 3.8 to 4.5. So most products have a good rating.

```python
sns.histplot(df['discount_percentage'], bins=20)
plt.title("Distribution of Discounted Percentage")
plt.xlabel("Discount")
plt.ylabel("Frequency")
```

Discount percentage is varying mostly cocentrating around the 40-60%.  

```python
sns.histplot(df['rating_count'], bins=30)
plt.title("Distribution of Rating Count")
plt.xlabel("Rating Count")
plt.ylabel("Frequency")
```

As expected, only a few products have a higher rating count (right skewed graph). That means, only a few products are famous and sold widely.
