## First importing the file

```python
import pandas as pd
df = pd.read_csv(r"amazon_dataset.csv")
df.head()
```

### Let's clean the data first. removing the '₹' sign:

```python
df['discounted_price'] = df['discounted_price'].replace('[₹,]', '', regex=True)
df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True)
```

### checking if cleared properly

```python
df['discounted_price'].head(5)
```

### removing '%' and ',':

```python
df['discount_percentage'] = df['discount_percentage'].replace('%', '', regex=True)
df['rating_count'] = df['rating_count'].replace(',', '', regex=True)
```

```python
df['discount_percentage'].head(5)
```

### Converting to proper types:

```python
df['product_id'] = df['product_id'].astype(str)
df['product_name'] = df['product_name'].astype(str)
df['category'] = df['category'].astype(str)

df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

df['about_product'] = df['about_product'].astype(str)
df['user_id'] = df['user_id'].astype(str)
df['user_name'] = df['user_name'].astype(str)
df['review_id'] = df['review_id'].astype(str)

df['review_title'] = df['review_title'].astype(str)
df['review_content'] = df['review_content'].astype(str)

df['img_link'] = df['img_link'].astype(str)
df['product_link'] = df['product_link'].astype(str)
```
